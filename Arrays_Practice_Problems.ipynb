Finding the union of 2 arrays using Hash_set:
1. Time complexity: O(n)
2. Space complexity: O(n)

```python
def find_union(a, b, c):
    hash_set = set()

    for i in range(len(a)):
        hash_set.add(a[i])

    for j in range(len(b)):
        hash_set.add(b[j])

    for _ in range(len(c)):
        hash_set.add(c[_])

    union_set = []
    for k in hash_set:
        union_set.append(k)

    return len(union_set)

a = [1, 2, 3, 4, 5]
b = [1, 2, 3]
c = [5, 3]
print(find_union(a, b, c))
```

Finding Intersection of 2 arrays:

```python
def find_intersection(a, b):
    res = []
    for i in range(len(b)):
        if b[i] in a:
            res.append(b[i])

    return res


a = [1, 7, -2, 3, 4, 5, 5, 3, 2]
b = [1, -2, 6, 7]
print(find_intersection(a, b))
```

```python
# you take inputs using the following way:
val = input("Enter elements separated by spaces: ").split()
arr = np.array(val, dtype=int)

x = int(input("X = "))

count = 0
for i in arr:
    if i==x:
        count+=1

print(f"The frequency of {x} is {count}.")
```

1. Time complexity: O(N)
2. Space complexity: O(N)

Better way of the above problem is as follows:

```python
x = int(input("X = "))
count = 0
for i in input("Enter elements separated by spaces: ").split():
    if int(i)==x:
        count+=1
print(count)
```

1. Time complexity: O(N)
2. Space complexity: O(1)
